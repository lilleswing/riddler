```python
import numpy as np
```

```python
def get_ink_std(distance):
    y = np.linspace(0, -distance, 1_000_000)
    
    high_x = np.cos(np.arcsin(y))
    high_x = 2 * np.nan_to_num(high_x, 0)
    
    low_x = np.cos(np.arcsin(distance + y))
    low_x = 2 * np.nan_to_num(low_x, 0)
    
    total_x = low_x + high_x
    return np.std(total_x)
```

```python
get_ink_std(1.692)
```

```python
get_ink_std(1.7)
```

```python
distances = np.linspace(1,2, 1000).tolist()
```

```python
stds = [get_ink_std(x) for x in distances]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,10))
plt.grid()
plt.xlabel("Distance")
plt.ylabel("Std Of Ink")
plt.plot(distances, stds)
```

```python
from scipy.optimize import minimize
```

```python
minimize(get_ink_std, 1.8)
```
