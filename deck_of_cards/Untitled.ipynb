```python
def play_game(l, debug=False):
    turns = 0
    while l[0] != 1:
        if debug:
            print(l)
        permute = l[0]
        sub1 = l[:permute]
        sub1.reverse()
        l = sub1 + l[permute:]
        turns += 1
    return turns
```

```python
def solve_it(deck_len):
    cards = list(range(1,deck_len+1))
    longest, best_order = 0, []
    for order in itertools.permutations(cards):
        order = list(order)
        game_len = play_game(order)
        if game_len > longest:
            longest = game_len
            best_order = [order]
        elif game_len == longest:
            best_order.append(order)
    return longest, best_order
```

```python
def get_moves(l):
    moves = []
    for index in range(1, len(l)):
        goal_value = index + 1
        if l[index] != goal_value:
            continue
        sub1 = l[:index+1]
        sub1.reverse()
        to_append = sub1 + l[index+1:]
        moves.append(to_append)
    return moves
        
```

```python
def reverse_dfs(l, depth):
    next_moves = get_moves(l)
    if len(next_moves) == 0:
        return 0, l
    best_depth, best_l = -1, None
    for move in next_moves:
        my_depth, my_l = reverse_dfs(move, depth+1)
        if my_depth > best_depth:
            best_depth, best_l = my_depth, my_l
    #print(l, best_l)
    return best_depth + 1, best_l
```

```python
reverse_dfs([1,2,3,4],0)
```

```python
import time
for i in range(11, 14):
    start = time.time()
    l = [x+1 for x in range(i+1)]
    retval = reverse_dfs(l, 0)
    end = time.time()
    print("%s,%s,%s,%s" % (i+1, retval[0], retval[1], end-start))
```
