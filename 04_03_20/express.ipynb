```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
_GRID_LINE_PROPERTIES = dict(color='#bdbdbd', linestyle='--', linewidth=0.5)
%matplotlib inline
```

```python
import sys
import numpy as np
import math
```

```python
steps_in_circle = int(30 / 3 * (math.pi * 3*2)/2)
steps_in_circle
```

```python
x = np.linspace(6, 3, 30)
p1 = [(-i,0) for i in x]
p2 = [(i,0) for i in x]

theta = np.linspace(0, np.pi, steps_in_circle)
r = 3
x = r*np.cos(theta)
y = r*np.sin(theta)
p1 = p1 + [(-i, j) for i, j in zip(x,y)]
p2 = p2 + [(i, -j) for i, j in zip(x,y)]

x = np.linspace(3, 6, 30)
p1 = p1 + [(i,0) for i in x]
p2 = p2 + [(-i,0) for i in x]
```

```python
def plot_people(ax, p1, p2):
    ax.scatter(p1[0], p1[1], c='blue')
    ax.scatter(p2[0], p2[1], c='orange')
    return ax
```

```python
def plot_by_index(i):
    if i % 20 == 0:
        print("index=%s" % i)
    return plot_people(ax, p1[i], p2[i])
```

```python
fig, ax = plt.subplots(figsize=(10,10))
fig.set_tight_layout(True)
ax.grid(**_GRID_LINE_PROPERTIES)

ax.set_xlim([-7,7])
ax.set_ylim([-7,7])

anim = FuncAnimation(fig, plot_by_index, frames=list(range(len(p1))), interval=1)
anim.save('line.gif', dpi=80)
plt.show()
#for p1_i, p2_i in zip(p1, p2):
#    plot_people(ax, p1_i, p2_i)
```
