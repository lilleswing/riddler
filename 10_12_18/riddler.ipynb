```python
import random
import math
def random_point():
    r = random.random()
    theta = random.random() * 2 * math.pi
    x = r * math.cos(theta)
    y = r * math.sin(theta)
    return x, y
points = []
n_points = 1000000
for i in range(n_points):
    points.append(random_point())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter([x[0] for x in points[:1000]], [x[1] for x in points[:1000]])
```

```python
import numpy as np
def distance(p1, p2):
    a = p1[0] - p2[0]
    b = p1[1] - p2[1]
    return (a*a + b*b) ** 0.5
    
def in_radius(points, r):
    n_in = 0
    for p in points:
        if distance(p, [0,1]) < r:
            n_in += 1
    return n_in /len(points)

distances = [distance(x, [1,0]) for x in points]
distances = np.array(distances)
```
