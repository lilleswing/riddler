```python
def A(r, x):
    return ((x - r) **2 + r**2 - x**2)**0.5
```

```python
def C(r, x):
    return ((-x - r) **2 + r**2 - x**2)**0.5
```

```python
def Y(r, x):
    return (r**2 - x**2)**0.5
```

```python
def is_int(i):
    return int(i) == i
```

```python
def find_values():
    r, r_max = 5, 100
    while r < r_max:
        x, x_max = 0.5, r
        while x < x_max:
            a = A(r,x)
            c = C(r,x)
            if is_int(a) and is_int(c):
                print(r, x)
                return r, x
            x += 0.5
        r += 0.5
r, x = find_values()
```

```python
points = [
    (r, 0),
    (x, Y(r,x)),
    (-x, Y(r,x)),
    (-r, 0),
    (-x, -Y(r,x)),
    (x, -Y(r,x))
    
]
```

```python
import matplotlib.pyplot as plt
_GRID_LINE_PROPERTIES = dict(color='#bdbdbd', linestyle='--', linewidth=0.5)
fig, ax = plt.subplots(figsize=(16,10))
ax.grid(**_GRID_LINE_PROPERTIES)

ax.scatter([x[0] for x in points], [x[1] for x in points])
for elem in points:
    ax.annotate(str(elem), elem)
```

```python
def distance(p1, p2):
    return ((p1[0] - p2[0]) ** 2 + (p2[1] - p1[1]) ** 2) ** 0.5
```

```python
for e1 in points:
    for e2 in points:
        d = distance(e1, e2)
        print(d, is_int(d))
```
