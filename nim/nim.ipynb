```python
starting_board = [1, 3, 5, 7]
```

```python
def xor_opt_fn(opt_fn):
    if opt_fn == max:
        return min
    return max

def construct_key(board, opt_fn):
    if opt_fn == max:
        return "%s,%s" % (board, 'max')
    return "%s,%s" % (board, 'min')

def get_next_moves(board):
    moves = []
    for i in range(len(board)):
        for j in range(1, board[i]+1):
            next_move = list(board)
            next_move[i] -= j
            moves.append(next_move)
    return moves

memoize = {}
def minimax(board, opt_fn):
    key = construct_key(board, opt_fn)
    if key in memoize:
        return memoize[key]
    if board == [0,0,0,0]:
        if opt_fn == max:
            return 1, []
        if opt_fn == min:
            return 0, []
    next_moves = get_next_moves(board)
    next_scores = []
    paths = []
    for move in next_moves:
        next_score, next_path = minimax(move, xor_opt_fn(opt_fn))
        next_scores.append(next_score)
        paths.append(next_path)
    index = next_scores.index(opt_fn(next_scores))
    my_score = next_scores[index]
    my_path = list(paths[index])
    my_path.append(next_moves[index])
    memoize[key] = my_score, my_path
    return my_score, my_path

minimax([1, 3, 5, 7], max)
```
